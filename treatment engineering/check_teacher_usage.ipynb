```python
import pandas as pd
from datetime import datetime
from tqdm import tqdm, trange
import scipy.stats as stats
```

```python
df_usage = pd.read_csv('raw_data/Teacher Usage - Time Series 2020-10-09T1635.csv')
```

```python
action_list = []
for i in trange(len(df_usage)):
    if df_usage.loc[i]['Event Type'] == 'Resource Downloaded':
        action_list.append('RD.' + df_usage.loc[i]['Curriculum Resource Category'])
    else:
        action_list.append(df_usage.loc[i]['Event Type'])
df_usage['action'] = action_list
df_usage['dow'] = [datetime.strptime(x[:10], '%Y-%m-%d').weekday() for x in df_usage['Usage Time']]

df_usage = df_usage[['dow','action']]
```

```python
friday_dic, non_friday_dic = {},{}
for i in tqdm(df_usage[df_usage['dow']==4]['action']):
    if i not in friday_dic.keys():
        friday_dic[i] = 1
    else:
        friday_dic[i] +=1

for i in tqdm(df_usage[df_usage['dow']!=4]['action']):
    if i not in non_friday_dic.keys():
        non_friday_dic[i] = 1
    else:
        non_friday_dic[i] +=1
```

```python
Keys = list(friday_dic.keys())
Keys.sort()
Keys.remove('User Session')
```

```python
friday_l, nfriday_l, diff_l = [],[],[]
for key in Keys:
    friday_l.append(friday_dic[key]/(sum(friday_dic.values()) - friday_dic['User Session']))
    nfriday_l.append(non_friday_dic[key]/(sum(non_friday_dic.values()) - non_friday_dic['User Session']))
diff_l = [x-y for x,y in zip(friday_l,nfriday_l)]
```

```python
df_sum = pd.DataFrame(
    {'Action': Keys,
     'Friday': friday_l,
     'Non Friday': nfriday_l,
     'Friday - Non Friday': diff_l
    })
```

```python
round(df_sum.sort_values(by=['Friday - Non Friday'],ascending=False),5)
```
